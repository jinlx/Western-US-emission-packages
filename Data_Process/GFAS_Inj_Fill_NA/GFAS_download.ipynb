```python
from ecmwfapi import ECMWFDataServer
```

```python
#years = ['2018','2019','2020']
months = ['01','02','03','04','05','06','07','08','09','10','11','12']
years = ['2017']
#months = ['07','08','09']
#months = ['08']
#years = ['2019','2020']
#months = ['01','02','03','04','05','06','07','08','09','10','11','12']
```

```python
# parameter can be checked in the web below
# https://confluence.ecmwf.int/display/CKB/CAMS%3A+Global+Fire+Assimilation+System+%28GFAS%29+data+documentation#CAMS:GlobalFireAssimilationSystem(GFAS)datadocumentation-ListofGFASparameters
for yy in years:
    for mm in months:
        print('Processing', yy+mm)
        print(yy + "-" + mm + "-" + "01" + "/to/" + yy + "-" + mm + "-31")
        server = ECMWFDataServer()
        server.retrieve({
            "class": "mc",
            "dataset": "cams_gfas",
            "date": yy + "-" + mm + "-" + "01" + "/to/" + yy + "-" + mm + "-31",
            "expver": "0001",
            "levtype": "sfc",
            #"param": "80.210/90.210/119.210/120.210/242.210/60.210", #CO2, OC, MAMI, ToP, BoP, inj_IS4FIRES
            # we don't have bop and inj_IS4FIRES before july 2018, thus delete these two fore data before the period.
            "param": "80.210/90.210/119.210/120.210", #CO2, OC, MAMI, ToP, BoP, inj_IS4FIRES
            "step": "0-24",
            "stream": "gfas",
            "target": "GFAS_" + yy + mm +"_hgts_raw.nc",
            "format": "netcdf",
            "time": "00",
            "type": "ga",
        })
```
